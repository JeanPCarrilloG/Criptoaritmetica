## Objetivo

- Considere un problema aritmético representado en letras:

    ### Condición
    - Asignar un dígito a cada letra
    
    ### Restricciones
    - Si una letra aparece más de una vez, debe recibir el mismo dígito
    - No es posible asignar el mismo dígito a más de una letra

![grafico.jpg](grafico.jpg)

### **Definición formal del problema**

* **Estado inicial**: Letras únicas sin números asociados.

* **Posibles acciones**: Valores numéricos que puede tomar cada letra  identifica.

* **Función de transiciones**: Relacionar un número a cada letra con un dígito del 0 al 9.

* **Prueba de satisfacción del objetivo**: Si los dígitos no se repiten y la relación con cada letra es única, además de cumplir la condición de operación suma, la prueba de satisfacción es efectiva.

* **Función de costo**: Cada acción tiene un costo de valor 0.

```python
problema = CriptoAritmetica(['send','more','money'])
estado = problema.estado_inicial
problema.pintar_estado(estado)
```

## Implementacion funciones de busqueda

- Función Backtracking

```python
%%time
busqueda = backtracking_search_aleatory_state(problema, estado)
print('Solución encontrada backtracking:', busqueda)
problema.pintar_estado(busqueda)
```

- Función Breadth First Search

```python
%%time
busqueda = breadth_first_aleatory_state(problema, estado)
print('Solución encontrada breadth-first:', busqueda)
problema.pintar_estado(busqueda)
```

- Función Best-first search

```python
%%time
busqueda = best_first_search_response_effective(problema, estado)
print("Solución encontrada best_first: ", busqueda)
problema.pintar_estado(dict(set(busqueda)))
```
